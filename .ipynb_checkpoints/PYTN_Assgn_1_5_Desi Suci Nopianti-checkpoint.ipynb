```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv('london_crime_by_lsoa.csv')
df.head()
```

## Data Cleaning
#### drop data agar lebih bersih dan tidak ada duplikat data

```python
df.drop('lsoa_code', axis=1, inplace=True)
df.head()
```

```python
print("----Informasi Baris dan Kolom----")
print("Baris : ", df.shape[0])
print("Kolom : ", df.shape[1])
```

## Duplikasi Data

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(inplace= False)
print("Jumlah duplikasi data : " + str(df.duplicated().sum()))
```

```python
print("----Informasi Baris dan Kolom----")
print("Baris : ", df.shape[0])
print("Kolom : ", df.shape[1])
```

## Area Plot

```python
df_sorted = df.sort_values(by=['year','month', 'borough'])
```

```python
crime_data_2016 = df[df['year'] == 2016]
borough_data = crime_data_2016.groupby('borough')['value'].sum().reset_index()

# plot the data using area plot
plt.figure(figsize=(70,20))
plt.stackplot(borough_data['borough'], borough_data['value'], labels=['Total Crime'], alpha=0.5)
plt.title('Total Crime in London Boroughs in 2016')
plt.xlabel('Boroughs')
plt.ylabel('Total Crime')
plt.legend(loc='upper left')

plt.show()
```

#### pada area plot ini menggambarkan jumlah kejahatan yang dilaporkan dalam tiap borought yang terjadi di tahun 2016

## Bar Chart

```python
crime_per_year = df.groupby('year')['value'].sum()
plt.figure(figsize=(20,5))
plt.bar(crime_per_year.index, crime_per_year.values)
plt.xlabel('Year')
plt.ylabel('Total Crime Value')
plt.title('Total Crime Value per Year')
plt.show()
```

##### pada grafik diatas menunjukan total kejahatan di loondon dari tahun 2008 sampai 2016. Dapat disimpulkan bahwa kejahatan terbanyak terjadi pada tahun 2012.

## Pie Chart
### grafik ini untuk memvisualisasikan proporsi tiap major kategori di seluruh borough setiap tahunnnya

```python
major_data = df.groupby('major_category')['value'].sum().reset_index()
plt.figure(figsize=(15,15))
plt.pie(major_data['value'], labels=major_data['major_category'], autopct='%1.2f%%', )
plt.axis('equal')
plt.show()
```

####  Dari pie chart di atas, kita dapat melihat bahwa Theft and Handling merupakan kategori kejahatan yang paling dominan di Westminster selama periode tahun 2008 hingga 2016, dengan persentase sebesar 36,4%. Sementara itu, kategori dengan persentase terendah adalah Sexual Offence, yang mencapai 0,0%. Informasi mengenai persentase kategori kejahatan lainnya juga tersedia dalam pie chart tersebut.

## Box Plot

```python
# Mengambil data major_category dan value dari dataframe
box_data = df[['major_category', 'value']]

# Membuat boxplot
plt.figure(figsize=(20,10))
plt.boxplot(box_data.groupby('major_category')['value'].apply(list), labels=box_data['major_category'].unique())
plt.title('Distribusi nilai value berdasarkan major category')
plt.ylabel('value')
plt.show()
```

#### Berdasarkan plot di atas, diketahui jika terdapat 8 kategori kejahatan di Westminster, yaitu Theft and Handling, Burglary, Criminal Damage, Drugs, dan Violence Against the Person, sexual offences, other notifiable offences, fraud of fegery . kategori other notifiable offence atau kategori kejahatan yang tidak termasuk dalam kategori-kategori kejahatan utama yang sering dilaporkan berada di tingkat paling tinggi.

## Word Clouds

```python
df_word = df.copy()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


relevant_data = df_word[df_word['year'] == 2016]['minor_category']


text = " ".join(relevant_data)


wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                min_font_size = 10).generate(text)

# Plot word cloud
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show() 
```
